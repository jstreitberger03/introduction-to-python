## 2. JSON Files

JSON (JavaScript Object Notation) is a common format for data exchange (APIs, configs). It looks exactly like a Python dictionary.

```python
import json

data = {
    "name": "FMA Course",
    "students": 20,
    "active": True
}

# Writing JSON
with open("data.json", "w") as f:
    json.dump(data, f)

# Reading JSON
with open("data.json", "r") as f:
    loaded_data = json.load(f)
    print(loaded_data)
    print(loaded_data["name"])
```

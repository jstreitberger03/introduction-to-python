# AI-assisted Programming

How large language models and coding assistants fit into a Python workflow.
        


## Learning Objectives

- Describe what AI coding assistants can and cannot do in a Python workflow.
- Write effective prompts that include context and plan to validate results with tests.
- Apply privacy and safety guardrails when using LLM-based tools.


## AI Coding Assistants

Tools like **GitHub Copilot** and **ChatGPT** are changing how we code.
Think of them as a "Junior Developer" sitting next to you.
- They are great at: Writing boilerplate code, explaining errors, and suggesting ideas.
- They are bad at: Understanding the "big picture" of your project or knowing your specific business rules.

## Prompt Engineering

The quality of the answer depends on the quality of your question.
- **Bad Prompt**: "Fix this code."
- **Good Prompt**: "This function is supposed to calculate tax, but it returns a negative number when the input is 0. Here is the error message..."

**Context is King**: The more information you give the AI (variable names, what you are trying to do), the better it can help.

## GitHub Copilot (example)
1. Install the Copilot extension in VS Code.
2. Sign in with GitHub and enable for your repositories.
3. Use inline suggestions or chat for explanations and quick snippets.
        


## Risks and Guardrails

**Don't trust the AI blindly.**
- **Hallucinations**: AI can confidently invent libraries or functions that don't exist.
- **Security**: Never paste passwords, API keys, or sensitive company data into a public AI chat.
- **Review**: Always read the code it generates. You are responsible for it, not the AI.

## Summary
AI tools speed up exploration and refactoring but require review, testing, and adherence to privacy policies.

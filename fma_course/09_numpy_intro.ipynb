```python
import numpy as np

# Creating an array
arr = np.array([1, 2, 3, 4, 5])
print(arr)
print(type(arr))
```

## 2. Operations

With lists, you need a loop to multiply every element by 2. With NumPy, you just multiply the array.

```python
# Element-wise multiplication
print(arr * 2)

# Element-wise addition
print(arr + 10)

# Mathematical functions
print(np.sqrt(arr))
```

## 3. 2D Arrays (Matrices)

NumPy is great for handling tables of numbers (matrices).

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix)
print("Shape:", matrix.shape)
```

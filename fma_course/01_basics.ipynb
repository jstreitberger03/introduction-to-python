## 2. Arithmetic Operations

Python can be used as a calculator. It supports standard mathematical operations:
- `+` Addition
- `-` Subtraction
- `*` Multiplication
- `/` Division
- `**` Power (Exponentiation)

```python
# Addition
print(3 + 4)

# Multiplication
print(5 * 6)

# Division (always returns a float)
print(10 / 2)

# Power
print(2 ** 3)  # 2 to the power of 3 is 8
```

## 3. Variables

Variables are like containers for storing data values. You can assign a value to a variable using the `=` sign.
Variable names should be descriptive (e.g., `price`, `name`, `total_amount`).

```python
x = 5
y = 10
result = x + y
print(result)
```

You can change the value of a variable at any time.

```python
price = 100
tax_rate = 0.2
tax = price * tax_rate
print("Tax:", tax)

price = 200
tax = price * tax_rate
print("New Tax:", tax)
```

## 4. Basic Data Types

Python has several built-in data types. The most common ones are:
- **Integers (`int`)**: Whole numbers (e.g., `5`, `-10`)
- **Floating Point Numbers (`float`)**: Decimal numbers (e.g., `3.14`, `0.001`)
- **Strings (`str`)**: Text (e.g., `"Hello"`, `'FMA'`)
- **Booleans (`bool`)**: True or False values (e.g., `True`, `False`)

You can check the type of a variable using the `type()` function.

```python
a = 10
print(type(a))

b = 3.14
print(type(b))

c = "Hello"
print(type(c))

d = True
print(type(d))
```

### Working with Strings
Strings can be combined (concatenated) using `+`.

```python
first_name = "Jane"
last_name = "Doe"
full_name = first_name + " " + last_name
print(full_name)
```

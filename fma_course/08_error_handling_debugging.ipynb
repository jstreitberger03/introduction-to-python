## 2. Try-Except Blocks

You can prevent your program from crashing by using `try` and `except` blocks.

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("You cannot divide by zero!")
    result = 0

print(f"Result is {result}")
```

## 3. Debugging Tips

1. **Read the Error Message**: Python tells you exactly what went wrong and where (line number).
2. **Print Debugging**: Insert `print()` statements to check variable values at different steps.
3. **Rubber Duck Debugging**: Explain your code line-by-line to an inanimate object (or a colleague). Often, you'll find the bug just by explaining it.

# Control Flow

In this section, we will learn how to control the flow of our program:
- **If/Else Statements**: Making decisions based on conditions.
- **Loops**: Repeating actions for every item in a list.

Control flow allows your program to make decisions and repeat actions. This is essential for automation.

## 1. Making Decisions (`if`, `elif`, `else`)

We use `if` statements to run code only when a certain condition is true.

### Real World Analogy
- **If** it is raining, take an umbrella.
- **Else if (elif)** it is cloudy, take a jacket.
- **Else**, wear a t-shirt.

```python
price = 150

if price > 100:
    print("This is expensive.")
else:
    print("This is affordable.")
```

### Comparison Operators
- `==` Equal to
- `!=` Not equal to
- `>` Greater than
- `<` Less than
- `>=` Greater than or equal to
- `<=` Less than or equal to

```python
department = "Legal"

if department == "IT":
    print("Contact Helpdesk")
elif department == "Legal":
    print("Contact General Counsel")
else:
    print("Contact HR")
```

## 2. Loops (`for`)

Loops allow you to repeat an action for every item in a list. This is incredibly powerful for processing data.
For example, imagine you have a list of 1000 filenames and you want to rename all of them. You would use a loop.

```python
names = ["Alice", "Bob", "Charlie", "David"]

for name in names:
    print("Hello, " + name + "!")
```

### Processing Data with Loops
Let's calculate the tax for a list of prices.

```python
prices = [10, 20, 50, 100]
tax_rate = 0.2

for price in prices:
    tax = price * tax_rate
    print("Price:", price, "Tax:", tax)
```

### Filtering Data
You can combine loops and `if` statements to filter data.

```python
transactions = [100, -50, 200, -20, 500]

print("Large transactions:")
for t in transactions:
    if t > 100:
        print(t)
```

## Summary
- Use `if` / `elif` / `else` to make decisions.
- Use `for` loops to process lists of data.
- Combine them to filter and analyze data.
